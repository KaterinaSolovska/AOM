```julia
function grad_w(X,y,w,b)
    return transpose(X)*(X*w + b - y)
end

function grad_b(X,y,w,b)
    g = 0
    
    for i=1:size(X)[1]
        g = g + sum(X[i,:].*w) + b[i] -y[i]
    end
    
    return g
end

function grad_descend(X,y,w0,b0, step_w, step_b)
    i = 0
    g_norm_w = 1.0    
    g_norm_b = 1.0
    
    while (i<5.0e6 && (g_norm_w > 0.05 || g_norm_b > 0.025))
        
        i=i+1
        g_w = grad_w(X,y,w0,b0)
        g_b = grad_b(X,y,w0,b0)
        
        w1 = w0 .- step_w.*g_w
        b1 = b0 .- step_b.*g_b
        
        w0 = w1
        b0 = b1
        
        g_norm_w = sqrt(sum(g_w.*g_w))
        g_norm_b = sqrt(sum(g_b.*g_b))

        
        if mod(i, 5.0e4)==0
            println("iter = ", i)
            println("  w = ", w0)
            println("g norm w = ", g_norm_w)
            println("g norm b = ", g_norm_b)
            println()
        end
    end 
    
    return w0, b0
end
```

```julia
# Pridan parametr b ~> y = b + X*w

b0 = rand(-10:10)*ones(size(X)[1])
w0 = rand(-1.0:0.0001:1.0, size(X)[2])

println("Initial estimation w = ", w0, ", b = ", b0[1])
w_final, b_final = grad_descend(X,y,w0,b0, 1.0e-8, 1.0e-6)

println("Heart diseas = ", b_final[1]," + (", w_final[1],")*biking + (", w_final[2],")*smoking")

fig = figure()
scatter(x1,y, s=10)
plot(x1_vis, b_final[1] .+ x1_vis.*w_final[1]+x2_1.*w_final[2], "g--", label="Smoking = 5.0%")
plot(x1_vis, b_final[1] .+ x1_vis.*w_final[1]+x2_2.*w_final[2], "y--", label="Smoking =15.0%")
plot(x1_vis, b_final[1] .+ x1_vis.*w_final[1]+x2_3.*w_final[2], "r--", label="Smoking = 25.0%")
xlabel("Biking")
ylabel("Heart disease")
title("Heart disease = "*string(round(b_final[1],digits=3))*" + ("*string(round(w_final[1],digits=3))*")*biking + ("* string(round(w_final[2],digits=3)) *")*smoking")


legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=2.0)
```
