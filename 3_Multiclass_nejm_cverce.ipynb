```julia
using PyPlot
using Distributions


# generovani 2D dat

mean1 = [2.,3.]
C = [0.5 0; 0 1.8]
d = MvNormal(mean1, C)
x1 = rand(d, 200)'

mean2 = [-1.5,2.75]
C2 = [1.9 0; 0 1.1]
d2 = MvNormal(mean2, C2)
x2 = rand(d2, 450)'

mean3 = [1.0,7.0]
C3 = [0.5 0; 0 0.8]
d3 = MvNormal(mean3, C3)
x3 = rand(d3, 200)'

y1 = zeros(size(x1)[1],3)
y2 = zeros(size(x2)[1],3)
y3 = zeros(size(x3)[1],3)

y1[:,1].=1
y2[:,2].=1
y3[:,3].=1

X = vcat(x1,x2,x3)
X = hcat(X, ones(size(X)[1]))

y = vcat(y1,y2,y3)

fig = figure()
PyPlot.scatter(x1[:,1],x1[:,2], s=10, label="Trida 1")
PyPlot.scatter(x2[:,1],x2[:,2], s=10, label="Trida 2")
PyPlot.scatter(x3[:,1],x3[:,2], s=10, label="Trida 3")


legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=2.0)

```

```julia
s1 = X[:,1]
s2 = X[:,2]

y1 = y[:,1]
y2 = y[:,2]
y3 = y[:,3]

# inicializace

w0 = rand(-10.0:0.01:10.0, 3,3)

w = inv(transpose(X)*X)*transpose(X)*y

val = X*w
class = zeros(size(X)[1])

m = findmax(val, dims=2)
for i=1:size(X)[1]
    class[i] = m[2][i][2]
end


######################################################################

fig = figure()
scatter(s1[class.==1],s2[class.==1] , s=10, label = "Klasifikovano jako Trida 1")
scatter(s1[class.==2],s2[class.==2] , s=10, label = "Klasifikovano jako Trida 2")
scatter(s1[class.==3],s2[class.==3] , s=10, label = "Klasifikovano jako Trida 3")
title("Klasifikace do 3 trid")
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=2.0)



```

```julia
# spatne klasifikovane body

ind_1 = (class.==1) .+ (y[:,1].==0)
ind_2 = (class.==2) .+ (y[:,2].==0)
ind_3 = (class.==3) .+ (y[:,3].==0)


fig = figure()
scatter(s1,s2, color="gray", s=6)
scatter(s1[ind_1.==2],s2[ind_1.==2] , s=10, label = "Nespravne klasifikovano jako Trida 1")
scatter(s1[ind_2.==2],s2[ind_2.==2] , s=10, label = "Nespravne klasifikovano jako Trida 2")
scatter(s1[ind_3.==2],s2[ind_3.==2] , s=10, label = "Nespravne klasifikovano jako Trida 3")
title("Chybne klasifikovane body")

legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=2.0)
```
